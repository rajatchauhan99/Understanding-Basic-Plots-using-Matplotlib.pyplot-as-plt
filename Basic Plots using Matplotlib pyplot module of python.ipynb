```python
x = [1,4,5,7,8]
y = [6,7,12,4,12]
```

```python
plt.plot(x,y)

plt.title("Graph between two variables")
plt.ylabel("Y axis")
plt.xlabel("X axis")
```

To plot multiple plots in the same output we use plt.subplot

```python
plt.subplot(211)

plt.plot(x,y, linewidth = 5)
plt.title("First Plot")

plt.subplot(212)

plt.bar(x,y)
plt.title("Second Plot")

# so here 2 1 1 means, 
# the first digit , 2 means we will have vertically 2 plots 
# and 1 means horizontally we will have one plot
# and last digit tells that in vertical position this plot will be the first plot.
```
